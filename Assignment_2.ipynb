```python
summury = df.describe()
max_values = df.max(numeric_only=True)
min_values = df.min(numeric_only=True)
mean_values = df.mean(numeric_only=True)
median_values = df.median(numeric_only=True)
std_div_values = df.std(numeric_only=True)
print(summury)
```

```python
print("Sum of Null (Missing) Values Per Column:")
null_counts = df.isnull().sum()
print(null_counts)
```

```python
#drop all null values.
df.dropna()

```

```python
print(" Starting Salary Analysis ---")

# a) Display records whose salary is greater than 50000.
print("a) \nRecords with Salary > 50000 ")
records_gt_50k = df[df['salary'] > 50000]
print(records_gt_50k)

# b) Display records whose salary is less than 20000.
print("b)\n Records with Salary < 20000")
records_lt_20k = df[df['salary'] < 20000]
print(records_lt_20k)

# c) Display records whose salary is between 20000 to 50000 .
print("\nc) Records with Salary between 20000 and 50000")
records_between_20_50k = df[df['salary'].between(20000, 50000)]
print(records_between_20_50k)

# d) Display records whose salary is greater than 100000.
print("\nd) Records with Salary > 100000")
records_gt_100k = df[df['salary'] > 100000]
print(records_gt_100k)

# e) Display records whose salary is less than 100000.
print("\n e) Records with Salary < 100000 ")
records_lt_100k = df[df['salary'] < 100000]
print(records_lt_100k)

```

```python
#a) display records whose discipline is A.
print("a)records whose discipline is A")
dis_A = df[df['discipline'] == 'A']
print(dis_A)
#b) display records whose discipline is B.
print("\na)records whose discipline is B")
dis_B = df[df['discipline'] == 'B']
print(dis_B)
#c) display records whose service is greater than 10.
print("\na)records whose service is greater than 10")
service_gt_10 = df[df['service'] > 10]
print(service_gt_10)
#d) display records whose service is less than 5.
print("\na)records whose service is less than 5")
service_ls_5 = df[df['service'] < 5]
print(service_ls_5 )

```

```python
# a) display records of phd and salary.
print("\na) records of phd and salary.")
records_ps = df[['phd','salary']]
print(records_ps) 

# b) display records of rank, sex and salary.
print("\nb) records of rank, sex and salary.")
records_rss = df[['rank', 'sex' ,'salary']]
print(records_rss) 

# c) display records without salary.
print("\nc) records without salary.")
records_wo_sal = df.drop(columns=['salary'])
print(records_wo_sal) 

# d) display first 50 records with rank,service and salary.
print("\nd) first 50 records with rank,service and salary.")
records_rss_50 = df[['rank', 'service' ,'salary']].head(50)
print(records_rss_50) 

# e) display records whose rank is Prof.
print("\ne) records whose rank is Prof.")
records_rank_prof = df[df['rank'] == 'Prof']
print(records_rank_prof) 

# f) display records whose rank is AssocProf.
print("\nf) records whose rank is AssocProf")
records_rank_AssocProf = df[df['rank'] == 'AssocProf']
print(records_rank_AssocProf) 
```
