```python
import numpy as np

x = np.array([2, 5, 8, 10])
y = np.array([3, 5, 6, 12])

print("Array x:", x)
print("Array y:", y)
print("x > y:", np.greater(x, y))
print("x >= y:", np.greater_equal(x, y))
print("x < y:", np.less(x, y))
print("x <= y:", np.less_equal(x, y))
```

```python
arr = np.arange(20, 51)
div_by_3 = arr[arr % 3 == 0]
remainder = np.remainder(div_by_3, 3)
print("Numbers divisible by 3:", div_by_3)
print("Element-wise remainder:", remainder)
```

```python
import pandas as pd

# i) Create a list of empids and names (10 employees)
emp_ids = [101, 102, 103, 104, 105, 106, 107, 108, 109, 110]
emp_names = ['Akshata', 'Neha', 'Rahul', 'Pooja', 'Suresh',
             'Kiran', 'Anita', 'Rohit', 'Priya', 'Vikas']

# ii) Convert list into Series
emp_series = pd.Series(emp_names, index=emp_ids)

# iii) Print type of Series
print("Type of emp_series:", type(emp_series))

# iv) Convert Series into DataFrame
emp_df = emp_series.reset_index()
emp_df.columns = ['EmpID', 'Name']

# v)
# a) Display all records
print("\nAll Records:")
print(emp_df)

# b) Display first five records
print("\nFirst five records:")
print(emp_df.head())

# c) Display last five records
print("\nLast five records:")
print(emp_df.tail())

# d) Display first two and last two records
print("\nFirst two records:")
print(emp_df.head(2))

print("\nLast two records:")
print(emp_df.tail(2))

# e) Display name of 5th employee
print("\nName of 5th employee:")
print(emp_df.loc[4, 'Name'])

# f) Display record of 4th employee
print("\nRecord of 4th employee:")
print(emp_df.loc[3])

# vi) Save the record in CSV file
emp_df.to_csv("MyRecord.csv", index=False)

print("\nData saved to MyRecord.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read mtcars.csv file
df = pd.read_csv("mtcars.csv")

# i) Line Plot (mpg vs hp)
plt.figure()
plt.plot(df['hp'], df['mpg'])
plt.xlabel("Horse Power")
plt.ylabel("Miles Per Gallon")
plt.title("Line Plot: HP vs MPG")
plt.show()

# ii)Bar Graph (mpg of first 5 cars)
plt.figure()
plt.bar(df.index[:5], df['mpg'][:5])
plt.xlabel("Car Index")
plt.ylabel("MPG")
plt.title("Bar Graph: MPG of first 5 cars")
plt.show()

# iii)pie Chart
plt.figure()
df['cyl'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Pie Chart: Cylinder Distribution")
plt.ylabel("")
plt.show()

# iv) Boxplot 
plt.figure()
plt.boxplot(df['mpg'])
plt.title("Boxplot of MPG")
plt.ylabel("MPG")
plt.show()

# v) Scatter Plot 
plt.figure()
plt.scatter(df['hp'], df['mpg'])
plt.xlabel("Horse Power")
plt.ylabel("MPG")
plt.title("Scatter Plot: HP vs MPG")
plt.show()

plt.figure()
plt.scatter(df['wt'], df['mpg'])
plt.xlabel("Weight")
plt.ylabel("MPG")
plt.title("Scatter Plot: Weight vs MPG")
plt.show()

# vii) Histogram (mpg)
plt.figure()
plt.hist(df['mpg'])
plt.xlabel("MPG")
plt.ylabel("Frequency")
plt.title("Histogram of MPG")
plt.show()
```
